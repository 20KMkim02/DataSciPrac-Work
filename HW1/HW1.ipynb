```python
import math
def BMI(weight: float,hight : float) -> str:
    weight=float(weight)
    hight=float(hight)
    if hight >3 :
        hight=hight/100
    elif hight <0:
        return 'false hight'
    BMI=round(weight/math.pow(hight,2),4)
    if BMI > 30:
        return (f'Your {BMI} is too much')
    elif BMI < 30 :
        return (f'Your {BMI} is normal')
w=input('Enter your weight : ')
h=input('Enter your hight : ')
BMI(w,h)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# all_movie={title:[reviewer_score,audience_score,released,[genres1,genres2,...]]}
# website ที่น่าสนใจ = https://www.2ebook.com/new/library/new_book/nlt
```

```python
# ใช้เวลารันประมาณ 1นาที
all_movie={'Moviename':['title','reviewer_score','audience_score','released_date',['genres']]}
# https://www.rottentomatoes.com/browse/movies_at_home/genres:action?page=2

genres = ["action", "adventure", "animation", "anime", "biography", 
          "comedy", "crime", "documentary", "drama", "entertainment", 
          "faith_and_spirituality", "fantasy", "game_show", "lgbtq+", 
          "health_and_wellness", "history", "holiday", "horror", "house_and_garden", 
          "kids_and_family", "music", "musical", "mystery_and_thriller", "nature", "news", 
          "reality", "romance", "sci-fi", "short", "soap", "special_interest", "sports", 
          "stand-up", "talk_show", "travel", "variety", "war", "western"]

# genres = ["action"]

for i in genres:
    url=f'https://www.rottentomatoes.com/browse/movies_at_home/genres:{i}?page=6'
    r= requests.get(url)
    data=r.text
    soup = BeautifulSoup(data, 'html.parser')
    # print(data)
    block=soup.find_all('div',{'class':'flex-container'})
    # print(len(block))
    for movie in block :
        movie_title=movie.find('a').find('span').get_text().strip()
        # print(movie_title)
        reviewer=movie.find('a').find('rt-text',{'slot':'criticsScore'}).get_text()
        # print(reviewer)
        audience=movie.find('a').find('rt-text',{'slot':'audienceScore'}).get_text()
        # print(audience)
        release=movie.find('a').find('span',{'class':'smaller'}).get_text().strip()
        dates = release.replace('Streaming',"")
        # print(dates)
        # print(movie_title,reviewer,audience,dates)
        if movie_title in all_movie:
            all_movie[movie_title][3].append(i)
            # print(movie_title) 
            # Above print is for checking more than 1 genres movie
        else :all_movie[movie_title]=[reviewer,audience,dates,[i]]
        
    # else :print(all_movie[movie_title][3].append(i))
print(all_movie)
```

<p>ทำการเปลี่ยนจาก dict -> ตาราง pandas

```python
all_movie.pop('Moviename')
df = pd.DataFrame.from_dict(all_movie, orient='index', columns=['reviewer_score', 'audience_score', 'released_date', 'Genres'])
df['Genres'] = df['Genres'].apply(lambda x: ", ".join(x))
# เปลี่ยน keyใน dict ให้เป็นอีก 1 column
df.reset_index(inplace=True)
df.rename(columns={'index': 'Title'}, inplace=True)
df
```
